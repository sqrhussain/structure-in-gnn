```python
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
university_dict = {}
classes_counters = {"all":{"student":0,"project":0,"course":0,"staff":0,"faculty":0}}
universities = ["cornell", "texas", "washington", "wisconsin"]


for university in universities:
    if university not in classes_counters:
        classes_counters[university] = {}
    with open('../data/graphs/raw/WebKB/' + university + '.content') as f: 
        for line in f:
            class_label = line.split()[-1]
            if class_label not in classes_counters[university]:
                classes_counters[university][class_label] = 0
            classes_counters[university][class_label] += 1
            classes_counters['all'][class_label] += 1
```

```python
plt.bar(list(classes_counters['cornell'].keys()), classes_counters['cornell'].values(), color='g')
plt.show()
```
